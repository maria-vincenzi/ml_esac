```python
def prepare_lc(filename):
    data=np.genfromtxt(filename, names=True, skip_header=22, skip_footer=1,
                   dtype=None, comments='DETECTION', usecols=[1,2,4,5])
    filter_list=['des'+i for i in data['FLT']]
    mjd_fromfirst=[d-data['MJD'][0] for d in data['MJD']]

    lc=Table(data, names=['general_mjd', 'single_filter','flux', 'flux_error'])
    zp=np.ones(len(data))*(27.5)
    zpsys=(['ab' for i in range(len(data))])
    lc.add_columns([Table.Column(name='mjd', data=mjd_fromfirst), \
                Table.Column(name='filter', data=filter_list), \
                Table.Column(name='zp', data=zp), \
                Table.Column(name='zpsys', data=zpsys)])
    return lc
```

```python
def peak_B_mag(model):
    times=np.arange(model.mintime(),model.maxtime(),0.5) 
    bandmags=((model.bandmag(band='bessellb', time=times, magsys='ab')))
    peak_mag=(np.min(bandmags[~np.isnan(bandmags)]))
    return peak_mag


def fit_supernova(lc, photo_z, photo_z_err, zspec, full=True, cov=True):
    """
    Small function to fit a light curve with the SALT2 model, using sncosmo and iminuit.
    
    Parameters
    -----------
    lc : astropy.table.Table
        Light curve (in the format sncosmo expects)
    
    Returns
    ----------
    z, t0, x0, x1, c
        Best-fitting parameters of the model
    """
    if full:
        bnds = {'z':(0.01, 1.5), 't0':(-100,100),'x0':(-1e-3, 1e-3), 'x1':(-3, 3), 'c':(-0.5, 0.5)}
        mod0 = sncosmo.Model('salt2-extended')
        res_nored = sncosmo.fit_lc(lc, mod0, vparam_names=mod0.param_names, bounds=bnds, minsnr=0)
        
        bnds = {'z':(photo_z-3*photo_z_err, photo_z+3*photo_z_err), 't0':(-100,100),'x0':(-1e-3, 1e-3), 'x1':(-3, 3), 'c':(-0.5, 0.5)}
        mod1 = sncosmo.Model('salt2-extended')
        res_phored = sncosmo.fit_lc(lc, mod1, vparam_names=mod1.param_names, bounds=bnds, minsnr=0)
        
        spec_err=0.0001
        bnds = {'z':(zspec-spec_err, zspec+spec_err), 't0':(-100,100),'x0':(-1e-3, 1e-3), 'x1':(-3, 3), 'c':(-0.5, 0.5)}
        mod2 = sncosmo.Model('salt2-extended')
        res_specred = sncosmo.fit_lc(lc, mod2, vparam_names=mod2.param_names, bounds=bnds, minsnr=0)
        if res_specred[0].success:
            mod2.parameters = res_specred[0].parameters
            COV_x1_c= res_specred[0].covariance[3,4]
            COV_x1_x0= res_specred[0].covariance[3,2]
            COV_c_x0= res_specred[0].covariance[2,4]
            if cov:
                return res_nored[0].parameters, res_nored[0].errors,\
                    res_phored[0].parameters, res_phored[0].errors,\
                    res_specred[0].parameters, res_specred[0].errors, COV_x1_c, COV_x1_x0, COV_c_x0
            else:
                return res_nored[0].parameters, res_nored[0].errors,\
                    res_phored[0].parameters, res_phored[0].errors,\
                    res_specred[0].parameters, res_specred[0].errors
        else:
            return None
    else:
        spec_err=0.0001
        bnds = {'z':(zspec-spec_err, zspec+spec_err), 't0':(-100,100),'x0':(-1e-3, 1e-3), 'x1':(-3, 3), 'c':(-0.5, 0.5)}
        mod2 = sncosmo.Model('salt2-extended')
        res_specred = sncosmo.fit_lc(lc, mod2, vparam_names=mod2.param_names, bounds=bnds, minsnr=0)
        if res_specred[0].success:
            z_fit, t0_fit, x0_fit, x1_fit, color_fit = res_specred[0].parameters
            deredshifted_model = sncosmo.Model('salt2-extended')
            deredshifted_model.parameters = [ 0, t0_fit, x0_fit, x1_fit, color_fit]
            mB = peak_B_mag(deredshifted_model)
            COV_x1_c= res_specred[0].covariance[3,4]
            COV_x1_x0= res_specred[0].covariance[3,2]
            COV_c_x0= res_specred[0].covariance[2,4]
            return res_specred[0].parameters, res_specred[0].errors, mB, COV_x1_c, COV_x1_x0, COV_c_x0
        else:
            return None
    
def to_log10(x, x_err):
    xlog= np.log10(x)
    xerr_log= 0.434*(x_err/x)
    return xlog, xerr_log
```
